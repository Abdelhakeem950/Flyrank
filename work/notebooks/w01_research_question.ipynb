## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*
**Lane:** Lane 4 - CTR / Engagement Opportunity Scoring

**Why this one:** I selected this lane because improving CTR for pages that already have high visibility offers a high return on investment. Instead of trying to rank new pages from scratch, we can capitalize on existing rankings by identifying pages that underperform relative to their specific position tier. This provides clear, actionable "quick wins" for the content team to update metadata or improve intent matching.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# قراءة البيانات مباشرة من رابط GitHub عشان نتفادى مشكلة مسارات Colab
url = 'https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# Define visible pages (e.g., page 1, >100 impressions)
visible_pages = df[(df['impressions_90d'] > 100) & (df['avg_position'] <= 10)]

# Define underperforming pages (e.g., CTR less than 3% despite being on page 1)
low_ctr_pages = visible_pages[visible_pages['ctr'] < 0.03]

print(f"Total highly visible pages (Page 1, >100 impressions): {len(visible_pages)}")
print(f"Visible pages with unusually low CTR (< 3%): {len(low_ctr_pages)}")
print(f"Opportunity size: {(len(low_ctr_pages)/len(visible_pages)*100):.1f}% of top pages could need metadata review.")
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**The Decision:** Deciding which highly visible pages should be prioritized for a metadata or content structure review.
**The Action:** An SEO specialist or content editor will review the flagged pages and rewrite the title tags, improve meta descriptions, or adjust the snippet structure to better capture clicks.
**Cost of a Wrong Call:** If the model recommends a page that is actually performing normally for its specific query type (a false positive), we waste the content team's limited time. Worse, if they edit a perfectly good title, we might accidentally decrease its current performance.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Quick look at average CTR by position tier to show why a tier-adjusted baseline is needed
tier_ctr = df[df['impressions_90d'] >= 100].groupby('position_tier')['ctr'].mean().sort_values(ascending=False)

print("Mean CTR by Position Tier (Showing why position context matters):")
print(tier_ctr.round(4).to_string())
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

To prove this lane is worth our time, let's look at pages ranking on Page 1 with meaningful volume (>=100 impressions). We will compare the CTR of the top 25% performing pages against the bottom 25%. The gap between these two groups represents our precise opportunity for improvement.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load the dataset using the direct URL
url = 'https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# Filter for pages on Page 1 with at least 100 impressions
page_1_data = df[(df['impressions_90d'] >= 100) & (df['position_tier'] == 'page_1')]

if not page_1_data.empty:
    top_quartile_ctr = page_1_data['ctr'].quantile(0.75)
    bottom_quartile_ctr = page_1_data['ctr'].quantile(0.25)

    print(f"Total eligible Page 1 URLs: {len(page_1_data)}")
    print(f"Top 25% of these pages achieve a CTR of at least: {top_quartile_ctr:.4f}")
    print(f"Bottom 25% of these pages suffer a CTR below: {bottom_quartile_ctr:.4f}")
    print(f"Conclusion: There is a massive gap in performance between the 75th and 25th percentile on Page 1, proving a strong opportunity for optimization.")
else:
    print("Dataset not loaded correctly or no data matches the criteria.")
```
